```python
titanic_data.drop('boat',axis = 1,inplace=True)
titanic_data.drop('cabin',axis = 1,inplace=True)
titanic_data.drop('body',axis = 1,inplace=True)
titanic_data.drop('home.dest',axis = 1,inplace=True)
```

```python
titanic_data.dropna(inplace=True)
```

```python
sns.heatmap(titanic_data.isnull(), yticklabels = False, cbar=False)
```

```python
titanic_data.isnull().sum()
# we can even fill the missing data with the mean value od that column
```

```python
titanic_data.head(2)
```

```python
# now we need a catagorical data as gender to be represneted in O and 1
# While applying logistic regression donot have a string values in a column we should pass a categorical data
sex = pd.get_dummies(titanic_data['sex'],drop_first = True ) #drop first because it give data of female as well as male
sex.head(5)
```

```python
embark = pd.get_dummies(titanic_data['embarked'], drop_first = True) 
#drop dirst column if next both columns are 0 means it was from C
embark
```

```python
Pcl = pd.get_dummies(titanic_data['pclass'], drop_first = True) 
#drop dirst column if next both columns are 0 means it was from 1
Pcl
```

```python
titanic_data = pd.concat([titanic_data, sex, embark, Pcl], axis=1)
titanic_data
```

```python
titanic_data.drop(['sex','passenger_id','embarked','name','ticket'], axis = 1, inplace = True)
titanic_data.drop(['pclass'], axis=1,inplace = True)
titanic_data
```

# Train Data

```python
X = titanic_data.drop(['survived'], axis = 1)
y = titanic_data['survived']
```

```python
#from sklearn.cross_validation import train_test_split
#above module was privously used but now it is there in below model
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
#LogisticRegression(solver='lbfgs')
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions) # it is to check a accuracy
```

```python
# check accuracy by confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
#check accuracy by accuracy_score method
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
